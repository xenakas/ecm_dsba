```r
data('food', package='PoEdata')
```

```r
load(file='food.rda')
```

```r
mod1<-lm(food_exp~income, data=food)
summary(mod1)
```

```r
plot(food$income,food$food_exp)
abline(mod1)
```

```r
plot(log(food$income),log(food$food_exp))
```

```r
w <- 1/food$income
mod2<-lm(food_exp~income, weight=w, data=food)
summary(mod1)
summary(mod2)
```
